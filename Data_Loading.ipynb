```python
from sklearn.preprocessing import LabelEncoder

df_lr = df.copy()
cols = ['proto', 'service', 'state']

label = LabelEncoder()

for col in cols:
    df_lr[col] = label.fit_transform(df_lr[col])

x = df_lr.drop(columns = ["attack_cat","label"])
y = df_lr["label"] 


from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,stratify=y random_state=42)

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)

model.fit(xtrain,ytrain)

ypred = model.predict(xtest)  

```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(ytest, ypred))

print("\nClassification Report:\n", classification_report(ytest, ypred))

print("\nConfusion Matrix:\n", confusion_matrix(ytest, ypred))
```

```python
a = df_lr.drop(columns=['attack_cat', 'label'])
b = df_lr['attack_cat']

from sklearn.model_selection import train_test_split
atrain, atest, btrain, btest = train_test_split(a, b,test_size=0.2,stratify=b,random_state=42)


from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100,random_state=42,n_jobs=-1)

rf_model.fit(atrain, btrain)
bpred = rf_model.predict(atest)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("Accuracy:", accuracy_score(btest, bpred))
print("\nClassification Report:\n", classification_report(btest, bpred))
print("\nConfusion Matrix:\n", confusion_matrix(btest, bpred))
```

```python
g = df_lr.drop(columns=['attack_cat', 'label'])
h = df_lr['attack_cat']

from sklearn.model_selection import train_test_split
gtrain, gtest, htrain, htest = train_test_split(g, h,test_size=0.2,stratify=h,random_state=42)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
gtrains = scaler.fit_transform(gtrain)
gtests = scaler.transform(gtest)

from sklearn.svm import SVC
svm_model = SVC(kernel='rbf',C=1.0,gamma='scale')

# # linear either this or upper one
# from sklearn.svm import LinearSVC
# svm_model = LinearSVC()

svm_model.fit(gtrains, htrain)
hpred = svm_model.predict(gtests)


from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


print("Accuracy:", accuracy_score(htest, hpred))
print("\nClassification Report:\n", classification_report(htest, hpred))
print("\nConfusion Matrix:\n", confusion_matrix(htest, hpred))
```
